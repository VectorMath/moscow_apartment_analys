## Общий анализ данных.

Теперь произведём общий анализ данных по квартирам чтобы получить информацию в целом. Нам нужно узнать следующее:

 * Среднее по:

Найти это нужно, группируя квартиры по:
 * **Типу квартиры.**
 * **Округам.**
 * **Количеству этажей в здание.**
 
Так же неплохо было бы посмотреть на количество квартир в зависимости от группировок.

### Средние значения квартир по типу жилья.

```python
# Средние значения по типу квартир.
means_by_type = df[['price_per_square_metr', 'price', 'square', 'distance_to_station', 'apartment_type']]\
                                                        .groupby('apartment_type').agg('mean').reset_index()
means_by_type = means_by_type.round(2)

# Количество квартир по типу жилья.
count_by_type = df[['price', 'apartment_type']].groupby('apartment_type').agg('count').reset_index()
count_by_type.rename({'price':'count'}, axis=1, inplace=True)

result = means_by_type.merge(count_by_type, how='inner', on='apartment_type')
result.sort_values('count', inplace=True)
result
```

```python
'''
Визуализация данных.
'''
fig, axes = plt.subplots(figsize=(14,7))

_ = result[['price','square']].plot.barh(ax=axes)
axes.set_yticklabels(result['apartment_type'])
axes.set_title('Гистограма средней цены и площади квартир по типу жилья', fontsize=18)
axes.legend(('Цена(млн рублей)','Площадь(метр квадратный)'))
plt.show()
```

```python
# Количество квартир по типу жилья.
result.sort_values('count', inplace=True)

fig, ax = plt.subplots(1, figsize=(14,7))
ax.barh(result['apartment_type'], result['count'])
ax.set_title('Количество квартир по типу жилья', fontsize=18)
ax.set_xlabel('Количество')
plt.show()
```

```python
# Средняя цена квартир по типу жилья.
result.sort_values('price', inplace=True)

fig, ax = plt.subplots(1, figsize=(14,7))
ax.barh(result['apartment_type'], result['price'])
ax.set_title('Средняя цена квартир по типу жилья', fontsize=18)
ax.set_xlabel('Млн рублей')
plt.show()
```

```python
# Средняя цена за квадратный метр по типу жилья.
result.sort_values('price_per_square_metr', inplace=True)

fig, ax = plt.subplots(1, figsize=(14,7))
ax.barh(result['apartment_type'], result['price_per_square_metr'])
ax.set_title('Средняя цена за квадратный метр по типу жилья.', fontsize=18)
ax.set_xlabel('Тысяча рублей')
plt.show()
```

```python
# Средняя площадь квартир по типу жилья.
result.sort_values('square', inplace=True)

fig, ax = plt.subplots(1, figsize=(14,7))
ax.barh(result['apartment_type'], result['square'])
ax.set_title('Средняя площадь квартир по типу жилья', fontsize=18)
ax.set_xlabel('Площадь м^2')
plt.show()
```

<div class="alert alert-info" role="alert">
  Исходя из общего анализа квартир по типу жилья, мы видим, что:
</div>

 * Больше всего предложений по:
     * **Квартира-студия**
     * **1-к квартира**
     * **2-к квартира**
     * **3-к квартира**

### Средние значения квартир по округам.

```python
# Средняя цена и площадь по округам.
means_by_district = df[['price_per_square_metr', 'price', 'square', 'distance_to_station', 'district']].groupby('district').agg('mean').reset_index()
means_by_district = means_by_district.round(2)

# Кол-во квартир по округам.
count_by_district = df[['price', 'district']].groupby('district').agg('count').reset_index()
count_by_district.rename({'price': 'count'}, axis=1, inplace=True)

# Добавим кол-во квартир к предыдущей таблице со средними значениями.
result = means_by_district.merge(count_by_district, how='inner', on='district')

# Вывод.
result
```

```python
'''
Визуализация данных.
'''
fig, axes = plt.subplots(figsize=(14,7))

_ = result[['price','square']].plot.barh(ax=axes)
axes.set_yticklabels(result['district'])
axes.set_title('Гистограма средней цены и площади квартир по округам', fontsize=18)
axes.legend(('Цена(млн рублей)','Площадь(метр квадратный)'))
plt.show()
```

```python
# Гистограмма количества квартир.

result.sort_values('count', inplace=True) # Сделаем сортировку, чтобы график был понятнее.

plt.barh(result['district'], result['count'])
plt.title('Количество квартир по округам', fontsize=18)
plt.show()
```

```python
# Гистограмма среднего пути до метро.

result.sort_values('distance_to_station', inplace=True) # Сделаем сортировку, чтобы график был понятнее.

plt.barh(result['district'], result['distance_to_station'])
plt.title('Средний путь до метро по округам', fontsize=18)
plt.xlabel('Путь в км')
plt.show()
```

```python
# Гистограмма средней цены квартиры по округам.

result.sort_values('price', inplace=True) # Сделаем сортировку, чтобы график был понятнее.

plt.barh(result['district'], result['price'])
plt.title('Средняя цена квартиры по округам', fontsize=18)
plt.xlabel('Млн рублей')
plt.show()
```

```python
# Гистограмма средней цены за квадратный метр по округам.

result.sort_values('price_per_square_metr', inplace=True) # Сделаем сортировку, чтобы график был понятнее.

plt.barh(result['district'], result['price_per_square_metr'])
plt.title('Средняя цена за один квадратный метр по округам', fontsize=18)
plt.xlabel('Тысяча рублей', fontsize=14)
plt.show()
```

<div class="alert alert-info" role="alert">
  Исходя из общего анализа квартир по округам, можно заметить, что:
</div>

 * В **Центральном округе** средняя цена на квартиру и её площадь самые большие.
 * В **Юго-восточном округе** средняя цена на квартиру и её площадь самые низкие.
 * Больше всего квартир продают в **Южном округе.**
 * Меньше всего квартир продают в **Северо-западном округе.**
 * Ближе всего к метро находятся квартиры в **Центральном** и **Западном** округе.
 * Дальше всех от метро находятся квартиры в **Юго-западном** и **Новомосковском** округе.
 
 
Меня заинтересовал средний путь до метро по округам.

### Средние значения квартир по количеству этажей в здание.

```python
# Средние значения.
means_by_total_floors = df.groupby('total_floors').agg('mean').reset_index()
means_by_total_floors = means_by_total_floors.round(2)

# Количество квартир.
count_by_total_floors = df.groupby('total_floors').agg('count').reset_index()
count_by_total_floors.drop(count_by_total_floors.columns[2:], axis=1, inplace=True)
count_by_total_floors.rename({'price_per_square_metr':'count'}, axis=1, inplace=True)

result = means_by_total_floors.merge(count_by_total_floors, how='inner', on='total_floors')
result
```

## Глубокий анализ в разрезе округа и типа недвижимости.

Мы получили общюю картину ситуации на рынке московской недвижимости, но хотелось бы увидеть что-то по конкретнее. Например сравнить цены или площадь однокомнатных квартир в разных округах. Сделаем это.

```python
# Сгруппируем наши данные.
detail_df = df.groupby(['apartment_type', 'district']).agg('mean')
detail_df = np.round(detail_df, 2)
detail_df.drop('total_floors', axis=1, inplace=True) # Уберём лишний столбец.

# Найдём кол-во записей в разрезе типа недвижимости и округа.
count = df[['price', 'apartment_type', 'district']].groupby(['apartment_type', 'district']).agg('count')
count.rename({'price': 'count'}, axis=1, inplace=True)

# Присоеденим к нашему датасету.
detail_df = detail_df.merge(count, how='inner', on=['apartment_type', 'district'])

# Посмотрим, что получилось.
detail_df
```
